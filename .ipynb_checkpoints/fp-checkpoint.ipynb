```python
import pickle
import pandas as pd

# Load the trained model
with open("flight_rf_latest.pkl", "rb") as file:
    model = pickle.load(file)

# Prepare a sample input data point
sample_input = {
    'Total_Stops': 1,
    'Journey_day': 24,
    'Journey_month': 3,
    'Dep_hour': 14,
    'Dep_min': 30,
    'Arrival_hour': 16,
    'Arrival_min': 45,
    'Duration_hours': 2,
    'Duration_mins': 15,
    'Air India': 0,
    'GoAir': 0,
    'IndiGo': 0,
    'Jet Airways': 1,
    'Jet Airways Business': 0,
    'Multiple carriers': 0,
    'Multiple carriers Premium economy': 0,
    'SpiceJet': 0,
    'Trujet': 0,
    'Vistara': 0,
    'Vistara Premium economy': 0,
    'Chennai': 0,
    'Delhi': 1,
    'Kolkata': 0,
    'Mumbai': 0,
    'Cochin': 1,
    'Delhi': 0,
    'Hyderabad': 0,
    'Kolkata': 0,
    'New Delhi': 0
}

# Convert the sample input to a DataFrame
input_df = pd.DataFrame([sample_input])

# Ensure the columns are in the correct order
input_df = input_df[['Total_Stops', 'Journey_day', 'Journey_month', 'Dep_hour',
                     'Dep_min', 'Arrival_hour', 'Arrival_min', 'Duration_hours',
                     'Duration_mins', 'Air India', 'GoAir', 'IndiGo',
                     'Jet Airways', 'Jet Airways Business',
                     'Multiple carriers', 'Multiple carriers Premium economy', 
                     'SpiceJet', 'Trujet', 'Vistara', 'Vistara Premium economy',
                     'Chennai', 'Delhi', 'Kolkata', 'Mumbai',
                     'Cochin', 'Delhi', 'Hyderabad', 'Kolkata', 'New Delhi']]

# Predict the price using the model
predicted_price = model.predict(input_df)

# Print the predicted price
print(f"The predicted flight price is: Rs. {predicted_price[0]:.2f}")
```
